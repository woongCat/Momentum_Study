- 데이터 전처리, 분류 모델, 회귀 모델 이렇게 배웠는데, 다들 잘 아시겠지만

- 머신러닝을 간단히 말해보면 패턴 파악입니다.

- 분류는 이런 데이터가 들어간다면 서쪽이고 다른 데이터가 들어가면 동쪽이다. 새로 들어온 이 데이터는 서쪽일까? 동쪽일까? 이분법적인 패턴을 찾는 방법입니다.
- 회귀는 이런 데이터가 들어간다면 이런 값이 나온다. 새로 데이터가 들어오면 이값일까? 회귀법적인 패턴을 찾는 방법입니다.

- 그러니까 여러분들은 데이터를 갖고 있을 때, 어떤 패턴이 나오는지 파악하는 연습을 한겁니다.
- 근데, 위와 같은 학습은 지도 학습인데, 어떤 분류가 나올지 알려주는 데이터가 있고, 회귀는 어떤 값이 결과인지 알려주는 데이터가 있습니다.

- 그런데, 거꾸로 라벨이 붙는 패턴을 알려주는 학습 방법이 있습니다.
- 바로 비지도 학습 방법입니다.

- 비지도 학습은 학습을 통해서 패턴을 찾는 게 아니라, 데이터를 통해서 패턴을 찾아내는 방법입니다.

그래서 오늘은 비지도학습 방법 2가지를 배워볼 예정인데, K-Means와 연관성 분석입니다.

그리고 오늘은 데이터를 보고 패턴을 찾아내는 연습을 할 예정입니다.

근데, 여러분이 패턴을 찾아내지 않고요 누군가가 만들어놓은 알고리즘을 이용해서 패턴을 찾을 예정입니다.
